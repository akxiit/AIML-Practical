```
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets


```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.concat([X, y], axis=1)
```

```
df.head()
```

```
df.info()
```

```
df.shape  ### shape of the dataset 150 rows and 5 columns
```

```
df.describe()
```

```
df.isna().sum()    ### Here not null value in the dataset
```

```
df['class'].value_counts()
```

```
sns.countplot(x='class', data=df, palette='Set2')        #### all class have same number of rows
plt.show()
```

### scatterplot between sepal Length and sepal Width

```
sns.scatterplot(x = df['sepal length'], y = df['sepal width'], hue=df['class'])
plt.show()
```
